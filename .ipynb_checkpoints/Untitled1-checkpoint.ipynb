### Cache Mechanism in the Neural Network

The cache in this neural network implementation is **a tuple of tuples**, designed to store essential data from the forward pass for use in the backward pass. This mechanism is critical for efficiently computing gradients during backpropagation.

#### Cache Structure

The cache consists of two main components:

- **Linear Cache (`linear_cache`)**: A tuple containing `(A_prev, W, b)`.
  - `A_prev`: Activations from the previous layer.
  - `W`: Weight matrix of the current layer.
  - `b`: Bias vector of the current layer.

- **Activation Cache (`activation_cache`)**: Stores `Z`, the output of the linear function before the activation function is applied.

#### Usage

During the forward pass, each layer's computations related to the linear function and the activation function are stored in `linear_cache` and `activation_cache`, respectively. This cached data is then unpacked in the backward pass to compute gradients:
